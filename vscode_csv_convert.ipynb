```python
import pandas as pd
import chardet
```

```python
# 1️⃣ 먼저 파일의 인코딩을 감지
def detect_encoding(file_path, sample_size=10000):
    with open(file_path, "rb") as f:
        rawdata = f.read(sample_size)  # 샘플 크기 지정 (너무 크면 속도 저하)
    return chardet.detect(rawdata)["encoding"]

# 2️⃣ 감지된 인코딩으로 CSV 파일 불러오기
file_path = "/content/data_added.csv"  # ⭐️ 여기에 파일 경로 입력
detected_encoding = detect_encoding(file_path)
print(f"📌 감지된 인코딩: {detected_encoding}")  # 예: 'ISO-8859-1', 'utf-16', 'cp949' 등

```

```python
import csv
# CSV 데이터를 리스트로 변환하는 함수
def csv_to_list(csv_text):
    data = []
    reader = csv.reader(csv_text.strip().split("\n"))
    for row in reader:
        data.append(row)
    return data
```
