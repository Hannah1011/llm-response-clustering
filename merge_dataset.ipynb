```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/human_feedback_ver2.csv", encoding='utf-8-sig')
df2 = pd.read_csv("/content/human_feedback_ver2_addrated.csv", encoding='utf-8-sig')
df3 = pd.read_csv("/content/humanfeedback_additional.csv", encoding = 'utf-8-sig')
```

```python
# ✅ 각 데이터셋의 행 개수 확인
print(f"📌 데이터셋 1 행 개수: {len(df1)}")
print(f"📌 데이터셋 2 행 개수: {len(df2)}")
print(f"📌 데이터셋 3 행 개수: {len(df3)}")
```

```python
# ✅ 데이터셋 병합 (열 개수가 다른 경우 자동으로 맞춰짐)
merged_df = pd.concat([df1, df2, df3], axis=0, ignore_index=True)

# ✅ 병합된 데이터셋 행 개수 확인
print(f"✅ 병합된 데이터셋 총 행 개수: {len(merged_df)}")
```

```python
# ✅ CSV 파일 저장
merged_df.to_csv("merged_human_feedback_ver2.csv", index=False, encoding="utf-8-sig")
print("✅ 병합된 CSV 파일 저장 완료!")
```
