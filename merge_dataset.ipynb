```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/human_feedback_ver2.csv", encoding='utf-8-sig') # 사용자 실제 데이터 (휴먼: 질의, 응답받음, 평가)
df2 = pd.read_csv("/content/humanfeedback_with_answer_csv_0221.csv", encoding='utf-8-sig') # 사용자 질의 및 답변 바탕 llm 가상 피드백 데이터 (휴먼: 질의, 응답받음 llm: 평가)
df3 = pd.read_csv("/content/humanfeedback_json_additional_0221.csv", encoding = 'utf-8-sig') # json 형태 사용자 질의 바탕 llm 답변 생성 및 가상 피드백 데이터 (휴먼: 질의 llm: 응답, 평가)
df4 = pd.read_csv("/content/humanfeedback_without_ans_csv_additional_0221.csv", encoding = 'utf-8-sig') # csv 형태 사용자 질의 바탕 llm 답변 생성 및 가상 피드백 데이터 (휴먼: 질의 llm: 응답, 평가)
```

```python
# ✅ 각 데이터셋의 행 개수 확인
print(f"📌 데이터셋 1 행 개수: {len(df1)}")
print(f"📌 데이터셋 2 행 개수: {len(df2)}")
print(f"📌 데이터셋 3 행 개수: {len(df3)}")
print(f"📌 데이터셋 4 행 개수: {len(df4)}")
```

```python
# ✅ 데이터셋 병합 (열 개수가 다른 경우 자동으로 맞춰짐)
merged_df = pd.concat([df1, df2, df3,df4], axis=0, ignore_index=True)

# ✅ 병합된 데이터셋 행 개수 확인
print(f"✅ 병합된 데이터셋 총 행 개수: {len(merged_df)}")
```

```python
# ✅ CSV 파일 저장
merged_df.to_csv("merged_human_feedback_ver3_0218.csv", index=False, encoding="utf-8-sig")
print("✅ 병합된 CSV 파일 저장 완료!")
```
